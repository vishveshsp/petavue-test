```python
import pandas as pd
from os import listdir, makedirs
from os.path import isfile, join, exists
import zipfile
import json
```

```python
airports = ['ATL', 'CLT', 'DEN', 'DFW', 'EWR', 'IAH', 'JFK', 'LAS', 'LAX', 'MCO', 'MIA', 'ORD', 'PHX', 'SEA', 'SFO']
flight_data = pd.DataFrame()
```

```python
for year in ['2016', '2017']:
    for month in range(1, 13):
        month_str = str(month).zfill(2)  # Pad month number with leading zero if needed
        folder_path = f"/Users/vishveshpal/Downloads/chrome and safari downloads/iitm downloads/New Folder With Items/{year}/On_Time_On_Time_Performance_{year}_{month}"
        files = [f for f in listdir(folder_path) if isfile(join(folder_path, f))]
        for file in files:
            if file.endswith('.csv'):
                df = pd.read_csv(join(folder_path, file))
                df_filtered = df[(df['Origin'].isin(airports)) & (df['Dest'].isin(airports))]
                flight_data = pd.concat([flight_data, df_filtered])
```

```python
weather_data = pd.DataFrame()

weather_files_path = '/Users/vishveshpal/Downloads/chrome and safari downloads/iitm downloads/New Folder With Items/'
for airport in airports:
    airport_folder_path = join(weather_files_path, airport)
    if not isfile(airport_folder_path):
        print('no file found')
        continue
    for year in range(2013, 2018):
        for month in range(1, 13):
            year_month = f"{year}-{str(month).zfill(2)}"
            file_path = join(airport_folder_path, f"{year_month}.json")
            if isfile(file_path):
                with open(file_path, 'r') as f:
                    weather_json = json.load(f)
                    df = pd.DataFrame(weather_json['data']['weather'])
                    df['airport'] = airport
                    weather_data = pd.concat([weather_data, df])
```

```python
merged_data = pd.merge(flight_data, weather_data, how='left', left_on=['FlightDate', 'Origin'], right_on=['date', 'airport'])
```
